# TESSERA encoder — input dimensionality & wrapper

**What it is:** a purpose-built agricultural pixel-timeseries SSL model (Cambridge). Its training weights
are gated, so we use the open **`geotessera`** product: precomputed *annual* 128-d embeddings (CC0).

**Role here:** `src/models/tessera.py` wraps it as `TesseraEncoder.encode(bench) -> (N, 128)`.

## Input contract — *not* a time series

Unlike Presto/OlmoEarth, TESSERA here is a **lookup**: one 128-d vector per `(lat, lon, year)`.
`encode` ignores the corrupted modality arrays entirely and keys only off `bench.latlon` (+ `year`).

**This makes it CONDITION-INVARIANT:** sensor-off / temporal-drop do **not** change its embedding,
so it anchors clean strict-transfer comparisons but cannot be stress-tested for sensor robustness.
That asymmetry is a finding to report, not a bug.

| input | from | notes |
|---|---|---|
| points | `bench.latlon` -> `(lon, lat)` | geotessera wants `(lon, lat)` order |
| `year` | `TESSERA_YEAR` (default 2024) | annual product; Benchmark has no per-sample year |
| **output** | `(N, 128)` | sampled from the precomputed tiles |

`encode` triggers network downloads of the 0.1° tiles covering the points, cached under `data/cache/`.

```python
# Bootstrap: find repo root and load a model module by path (src has no __init__ files).
# Use a kernel that has the deps (numpy, torch, and the model's package).
import sys, importlib.util
from pathlib import Path
from types import SimpleNamespace
import numpy as np

REPO = Path.cwd()
while not (REPO / 'src').exists():
    REPO = REPO.parent
sys.path.insert(0, str(REPO))

def load(name, rel):
    spec = importlib.util.spec_from_file_location(name, REPO / rel)
    m = importlib.util.module_from_spec(spec); sys.modules[name] = m; spec.loader.exec_module(m)
    return m
```

```python
tessera = load('tessera_mod', 'src/models/tessera.py')
print('embedding dim:', tessera.TESSERA_EMBEDDING_DIM)
print('default year :', tessera.TESSERA_YEAR)
print('cache dir    :', tessera.DEFAULT_CACHE_DIR)
```

```python
# A tiny synthetic Benchmark (duck-typed): just the attributes an encoder reads.
# This matches what src/dataio/get_input.py produces, but needs no data on disk.
N, T = 4, 12
rng = np.random.default_rng(0)
bench = SimpleNamespace(
    s2=(rng.random((N, T, 11)) * 5000).astype('float32'),   # B2..B12, NDVI (reflectance-ish)
    s1=(rng.random((N, T, 2)) * -15).astype('float32'),     # VV, VH (dB-ish)
    climate=(rng.random((N, T, 3)) * [300, 0.01, 500]).astype('float32'),  # temp(K), precip(m), elev(m)
    s2_mask=np.ones((N, T), 'float32'),
    s1_mask=np.ones((N, T), 'float32'),
    climate_mask=np.ones((N, T), 'float32'),
    doy=np.tile(np.linspace(15, 350, T), (N, 1)).astype('float32'),
    latlon=np.array([[11.0, 79.0]] * N, 'float32'),  # Cauvery Delta-ish (lat, lon)
    s2_bands=['B2','B3','B4','B5','B6','B7','B8','B8A','B11','B12','NDVI'],
    s1_bands=['VV','VH'],
    climate_bands=['temperature','precipitation','elevation'],
)
print('synthetic bench:', N, 'samples x', T, 'timesteps')
```

## Condition-invariance (no network needed to see the design)
`encode` reads only `latlon`; corrupting the bench changes nothing. We show the wrapper turns
`latlon=[lat,lon]` into geotessera's `(lon,lat)` points.

```python
lat, lon = bench.latlon[:, 0], bench.latlon[:, 1]
points = [(float(lo), float(la)) for la, lo in bench.latlon]  # (lon, lat)
print('latlon[0] = [lat,lon] =', bench.latlon[0], '-> point (lon,lat) =', points[0])
```

## Caveats
- TESSERA is **condition-invariant** here -> exclude it from sensor-robustness claims; use it for clean transfer.
- The product is annual; we pick a single representative `year` (Benchmark has no per-sample year).
- First run downloads tiles (tens of MB per 0.1° cell) into `data/cache/`.
