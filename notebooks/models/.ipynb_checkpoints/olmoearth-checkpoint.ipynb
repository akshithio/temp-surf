# OlmoEarth encoder — input dimensionality & wrapper

**What it is:** a strong general-purpose multimodal EO foundation model (Ai2). It is a *spatial* model
(2.56 km chips), but its own crop-timeseries eval (BreizhCrops) runs it on single pixels — which is exactly
what we do.

**Role here:** `src/models/olmoearth.py` wraps it frozen: `OlmoEarthEncoder.encode(bench) -> (N, D)`,
where `D = 768 * (#modalities)` for Base (S2 + S1 -> **1536**).

## Input contract (`MaskedOlmoEarthSample`, BHWTC layout)

| field | shape | meaning |
|---|---|---|
| `sentinel2_l2a` | `(B, H, W, T, 12)` | 12 S2 L2A bands |
| `sentinel2_l2a_mask` | `(B, H, W, T, 3)` | per **band-set** (3 sets by resolution) |
| `sentinel1` | `(B, H, W, T, 2)` | `vv, vh` |
| `sentinel1_mask` | `(B, H, W, T, 1)` | 1 band-set |
| `timestamps` | `(B, T, 3)` | `[day, month(0-idx), year]` per timestep |

We use `H = W = 1` (single pixel) and `patch_size = 1`. S2 L2A order is
`[B02,B03,B04,B08,B05,B06,B07,B8A,B11,B12,B01,B09]` in 3 band-sets:
`{B02,B03,B04,B08}` (10m), `{B05,B06,B07,B8A,B11,B12}` (20m), `{B01,B09}` (60m).
Our `Benchmark` lacks the 60m set `{B01,B09}` -> that band-set is marked **MISSING**.
Inputs are normalized with the pretrained `Normalizer(Strategy.COMPUTED)`.
Stress conditions propagate through the masks: a missing modality/timestep becomes `MaskValue.MISSING`,
so sensor-off / temporal-drop genuinely change the embedding (unlike TESSERA).

```python
# Bootstrap: find repo root and load a model module by path (src has no __init__ files).
# Use a kernel that has the deps (numpy, torch, and the model's package).
import sys, importlib.util
from pathlib import Path
from types import SimpleNamespace
import numpy as np

REPO = Path.cwd()
while not (REPO / 'src').exists():
    REPO = REPO.parent
sys.path.insert(0, str(REPO))

def load(name, rel):
    spec = importlib.util.spec_from_file_location(name, REPO / rel)
    m = importlib.util.module_from_spec(spec); sys.modules[name] = m; spec.loader.exec_module(m)
    return m
```

```python
olmo = load('olmo_mod', 'src/models/olmoearth.py')
from olmoearth_pretrain.data.constants import Modality
s2, s1 = Modality.SENTINEL2_L2A, Modality.SENTINEL1
print('S2 L2A band_order:', s2.band_order)
print('S2 band-sets:', [(bs.bands, bs.resolution_factor) for bs in s2.band_sets])
print('S1 band_order:', s1.band_order, '| sizes ->', olmo.SIZE_TO_MODEL_ID)
```

```python
# A tiny synthetic Benchmark (duck-typed): just the attributes an encoder reads.
# This matches what src/dataio/get_input.py produces, but needs no data on disk.
N, T = 4, 12
rng = np.random.default_rng(0)
bench = SimpleNamespace(
    s2=(rng.random((N, T, 11)) * 5000).astype('float32'),   # B2..B12, NDVI (reflectance-ish)
    s1=(rng.random((N, T, 2)) * -15).astype('float32'),     # VV, VH (dB-ish)
    climate=(rng.random((N, T, 3)) * [300, 0.01, 500]).astype('float32'),  # temp(K), precip(m), elev(m)
    s2_mask=np.ones((N, T), 'float32'),
    s1_mask=np.ones((N, T), 'float32'),
    climate_mask=np.ones((N, T), 'float32'),
    doy=np.tile(np.linspace(15, 350, T), (N, 1)).astype('float32'),
    latlon=np.array([[11.0, 79.0]] * N, 'float32'),  # Cauvery Delta-ish (lat, lon)
    s2_bands=['B2','B3','B4','B5','B6','B7','B8','B8A','B11','B12','NDVI'],
    s1_bands=['VV','VH'],
    climate_bands=['temperature','precipitation','elevation'],
)
print('synthetic bench:', N, 'samples x', T, 'timesteps')
```

## Wrapper mapping: `Benchmark -> OlmoEarth chip`
`_build_modality` maps each Benchmark band (case-insensitive: S1 is `vv`/`vh`) into OlmoEarth's order,
normalizes, and builds the per-band-set mask. The `{B01,B09}` set has no source -> stays MISSING.

```python
enc = olmo.OlmoEarthEncoder(size='base', include_s1=True, device='cpu', batch_size=4)
enc._ensure_loaded()  # loads Base weights from HuggingFace cache
s2_img, s2_mask = enc._build_modality(bench.s2, bench.s2_bands, bench.s2_mask, enc._modality.SENTINEL2_L2A)
print('S2 image (N,1,1,T,12):', s2_img.shape)
print('S2 mask  (N,1,1,T,3) :', s2_mask.shape)
miss = int(enc._maskval.MISSING.value)
print('60m band-set [B01,B09] all MISSING:', bool((s2_mask[:, 0, 0, :, 2] == miss).all()))
```

## Forward pass → embedding
Returns `(N, 1536)` for Base (768 S2 ⊕ 768 S1).

```python
emb = enc.encode(bench)
print('embedding:', emb.shape, '| dim:', enc.embedding_dim, '| finite:', bool(np.isfinite(emb).all()))
```

## Caveats
- OlmoEarth is a spatial model used at 1x1 pixels — an intentionally 'unnatural' use, matching its own
  BreizhCrops crop-timeseries eval.
- We lack the 60m band-set `{B01,B09}` and have no Landsat -> those are MISSING.
- `embedding_dim` is only known after the first forward pass (it's `-1` until then).
