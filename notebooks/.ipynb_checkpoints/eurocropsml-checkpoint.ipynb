# EuroCropsML EDA

Exploratory data analysis of the EuroCropsML benchmark.

**Benchmark**: Multiclass crop-type classification, ~706k parcels across 3 countries
**Bands**: 10 S2 bands (B2..B8A, B11, B12) + NDVI (computed)
**Timesteps**: Variable (36–60 irregular dates per parcel, resampled to fixed length)
**Split**: Train on Latvia + Portugal → test on Estonia (transnational transfer)

```python
%load_ext autoreload
%autoreload 2

import sys
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

REPO = Path.cwd().resolve().parent
sys.path.insert(0, str(REPO))

from src.io.get_input import load_eurocropsml

sns.set_style("whitegrid")
plt.rcParams["figure.dpi"] = 120
```

## Sampling from raw NPZ files

First, inspect raw NPZ metadata to understand time-series length distribution, date ranges, and country breakdown.

```python
preprocess_dir = REPO / "data" / "input" /  "eurocropsml" / "preprocess"
all_npz = sorted(preprocess_dir.glob("*.npz"))
print(f"Total NPZ files: {len(all_npz):,d}")

countries = {"EE": 0, "LV": 0, "PT": 0}
for p in all_npz:
    prefix = p.stem[:2]
    if prefix in countries:
        countries[prefix] += 1
    else:
        countries[prefix] = countries.get(prefix, 0) + 1

country_names = {"EE": "Estonia", "LV": "Latvia", "PT": "Portugal"}
for code, count in sorted(countries.items()):
    name = country_names.get(code, code)
    print(f"  {name} ({code}): {count:>7,d} ({count/len(all_npz):>5.1%})")
```

```python
rng = np.random.default_rng(42)
sample_files = rng.choice(all_npz, size=min(5000, len(all_npz)), replace=False)

ts_lengths = []
date_starts = []
date_ends = []
countries_sampled = []

for p in sample_files:
    with np.load(str(p)) as data:
        t = data["data"].shape[0]
        dates = data["dates"]
    ts_lengths.append(t)
    date_starts.append(dates[0])
    date_ends.append(dates[-1])
    countries_sampled.append(p.stem[:2])

ts_lengths = np.array(ts_lengths)
date_starts = np.array(date_starts, dtype="datetime64[D]")
date_ends = np.array(date_ends, dtype="datetime64[D]")

print("Time series lengths:")
print(f"  min:    {ts_lengths.min()}")
print(f"  max:    {ts_lengths.max()}")
print(f"  median: {np.median(ts_lengths):.0f}")
print(f"  mean:   {ts_lengths.mean():.1f}")
print(f"  std:    {ts_lengths.std():.1f}")
print(f"\nDate range:")
print(f"  Start: {pd.Timestamp(date_starts.min()).strftime('%Y-%m-%d')}")
print(f"  End:   {pd.Timestamp(date_ends.max()).strftime('%Y-%m-%d')}")
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

axes[0].hist(ts_lengths, bins=50, color="#3498db", edgecolor="white")
axes[0].axvline(ts_lengths.min(), color="#e74c3c", ls="--", label=f"min={ts_lengths.min()}")
axes[0].axvline(ts_lengths.max(), color="#27ae60", ls="--", label=f"max={ts_lengths.max()}")
axes[0].set_xlabel("Timesteps"); axes[0].set_ylabel("Count")
axes[0].set_title("Time series length distribution"); axes[0].legend()

countries_arr = np.array(countries_sampled)
for i, code in enumerate(["EE", "LV", "PT"]):
    mask = countries_arr == code
    if mask.sum() > 0:
        axes[1].hist(ts_lengths[mask], bins=30, alpha=0.6, label=country_names[code])
axes[1].set_xlabel("Timesteps"); axes[1].set_ylabel("Count")
axes[1].set_title("Length by country"); axes[1].legend()

for i, code in enumerate(["EE", "LV", "PT"]):
    mask = countries_arr == code
    if mask.sum() > 0:
        axes[2].scatter(date_starts[mask], ts_lengths[mask], s=3, alpha=0.3, label=country_names[code])
axes[2].set_xlabel("Start date"); axes[2].set_ylabel("Timesteps")
axes[2].set_title("Length vs start date"); axes[2].legend(markerscale=5)

plt.suptitle("Raw NPZ time-series characteristics", fontsize=14, y=1.02)
plt.tight_layout()
```

## Load the Benchmark (sampled subset)

Loading all 706k parcels is expensive. We use a 30k subset for EDA. The loader resamples variable-length series to 96 timesteps.

```python
bench = load_eurocropsml(
    root=REPO / "data" / "input",
    timesteps=96, max_samples=30000, shuffle=True, seed=0,
)
print(f"Samples:   {bench.n_samples:,d}")
print(f"Timesteps: {bench.timesteps}")
print(f"Task:      {bench.task}")
print(f"S2 shape:  {bench.s2.shape}")
print(f"S2 bands ({len(bench.s2_bands)}): {bench.s2_bands}")
print(f"S1 shape:  {bench.s1.shape} (always empty -- EuroCropsML has no S1)")
print(f"Climate:   {bench.climate.shape} (always empty)")
```

## Country and Region Distribution

```python
groups, counts = np.unique(bench.groups, return_counts=True)
group_df = pd.DataFrame({"country": groups, "count": counts}).sort_values("count", ascending=False)

colors = {"Estonia": "#3498db", "Latvia": "#e67e22", "Portugal": "#27ae60"}

fig, ax = plt.subplots(figsize=(7, 4))
bar_colors = [colors.get(g, "#95a5a6") for g in group_df["country"]]
bars = ax.bar(group_df["country"], group_df["count"], color=bar_colors)
for bar in bars:
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 200,
            f"{int(bar.get_height()):,d}", ha="center", fontweight="bold")
ax.set_ylabel("Samples")
ax.set_title("EuroCropsML samples per country (30k subset)")
plt.tight_layout()

print(f"Number of unique countries/groups: {len(groups)}")
for g in groups:
    mask = bench.groups == g
    n_classes = len(np.unique(bench.labels[mask]))
    print(f"  {g}: {mask.sum():,d} samples, {n_classes} crop-type classes")
```

```python
class_counts = pd.Series(bench.labels).value_counts().head(20)
label_map = {i: bench.label_names[i] for i in range(len(bench.label_names))}
class_names = [label_map.get(i, f"class_{i}") for i in class_counts.index]

fig, ax = plt.subplots(figsize=(12, 5))
ax.bar(range(len(class_counts)), class_counts.values, color="#3498db")
ax.set_xticks(range(len(class_counts)))
ax.set_xticklabels(class_names, rotation=90, fontsize=7)
ax.set_ylabel("Count")
ax.set_title("Top 20 crop-type classes (by sample count)")
plt.tight_layout()

print(f"Total unique crop types in subset: {len(np.unique(bench.labels))}")
print(f"Total available in full dataset:   {len(bench.label_names)}")
```

## NDVI Temporal Profiles

```python
ndvi_idx = bench.s2_bands.index("NDVI")
ndvi_mean = bench.s2[:, :, ndvi_idx].mean(axis=0)
ndvi_std = bench.s2[:, :, ndvi_idx].std(axis=0)
doy_mean = bench.doy.mean(axis=0)

fig, axes = plt.subplots(1, 2, figsize=(14, 4))

axes[0].plot(doy_mean, ndvi_mean, "-o", color="#27ae60", ms=3)
axes[0].fill_between(doy_mean, ndvi_mean - ndvi_std, ndvi_mean + ndvi_std, alpha=0.15, color="#27ae60")
axes[0].set_xlabel("Day of year"); axes[0].set_ylabel("NDVI")
axes[0].set_title("Mean NDVI profile (std), all countries")

for g in groups:
    mask = bench.groups == g
    ndvi_g = bench.s2[mask, :, ndvi_idx].mean(axis=0)
    axes[1].plot(doy_mean, ndvi_g, "-", label=g, color=colors.get(g))
axes[1].set_xlabel("Day of year"); axes[1].set_ylabel("NDVI")
axes[1].set_title("Mean NDVI per country"); axes[1].legend()

plt.suptitle("EuroCropsML temporal NDVI profiles", fontsize=14, y=1.02)
plt.tight_layout()
```

```python
mask_mean = bench.s2_mask.mean(axis=0)

fig, ax = plt.subplots(figsize=(8, 3))
ax.fill_between(doy_mean, mask_mean, alpha=0.5, color="#3498db")
ax.plot(doy_mean, mask_mean, "-o", color="#2c3e50", ms=3)
ax.set_xlabel("Day of year"); ax.set_ylabel("Fraction available")
ax.set_ylim(-0.05, 1.05)
ax.set_title("Per-timestep observation availability (after resampling to 96 steps)")
plt.tight_layout()
```

## Band Reflectance Statistics

```python
band_stats = []
for ci, band in enumerate(bench.s2_bands):
    vals = bench.s2[:, :, ci][bench.s2_mask > 0]
    band_stats.append({"band": band, "mean": vals.mean(), "std": vals.std(),
        "p05": np.percentile(vals, 5), "p25": np.percentile(vals, 25),
        "p50": np.percentile(vals, 50), "p75": np.percentile(vals, 75), "p95": np.percentile(vals, 95)})
pd.DataFrame(band_stats)
```

```python
spectral_bands = [b for b in bench.s2_bands if b.startswith("B")]
spec_idx = [bench.s2_bands.index(b) for b in spectral_bands]

fig, ax = plt.subplots(figsize=(10, 4))
for g in groups:
    mask = bench.groups == g
    mean_spec = bench.s2[mask, :, :][:, :, spec_idx].mean(axis=(0, 1))
    ax.plot(spectral_bands, mean_spec, "-o", label=g, color=colors.get(g), ms=5)
ax.set_xlabel("Band"); ax.set_ylabel("Mean reflectance")
ax.set_title("Mean spectral profile per country")
ax.legend()
plt.tight_layout()
```

## Spatial Distribution (Lat / Lon)

```python
latlon = bench.latlon
valid = np.isfinite(latlon[:, 0]) & np.isfinite(latlon[:, 1])

fig, ax = plt.subplots(figsize=(8, 6))
for g in groups:
    mask = (bench.groups == g) & valid
    ax.scatter(latlon[mask, 1], latlon[mask, 0], s=2, alpha=0.5, label=g, color=colors.get(g))
ax.set_xlabel("Lon"); ax.set_ylabel("Lat")
ax.set_title("Geographic distribution (30k subset)")
ax.legend(markerscale=5)
plt.tight_layout()
```

## Transnational Split Analysis

The evaluation split is Latvia + Portugal → Estonia. Let's inspect the domain shift.

```python
train_mask = bench.groups != "Estonia"
test_mask = bench.groups == "Estonia"

train_classes = set(bench.labels[train_mask])
test_classes = set(bench.labels[test_mask])
unseen = test_classes - train_classes

print(f"Train samples (Latvia+Portugal): {train_mask.sum():,d}")
print(f"Test samples  (Estonia):         {test_mask.sum():,d}")
print(f"Unique classes in train:         {len(train_classes)}")
print(f"Unique classes in test:          {len(test_classes)}")
print(f"Classes in test NOT in train:    {len(unseen)}")
if unseen:
    names = [bench.label_names[int(i)] for i in sorted(unseen)]
    print(f"  e.g.: {", ".join(names[:10])}")

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

ndvi_train = bench.s2[train_mask, :, ndvi_idx].mean(axis=0)
ndvi_test = bench.s2[test_mask, :, ndvi_idx].mean(axis=0)
axes[0].plot(doy_mean, ndvi_train, "-", color="#e67e22", label="Train (LV+PT)")
axes[0].plot(doy_mean, ndvi_test, "-", color="#3498db", label="Test (EE)")
axes[0].set_xlabel("Day of year"); axes[0].set_ylabel("NDVI")
axes[0].set_title("Mean NDVI: train vs test"); axes[0].legend()

spec_train = bench.s2[train_mask, :, :][:, :, spec_idx].mean(axis=(0, 1))
spec_test = bench.s2[test_mask, :, :][:, :, spec_idx].mean(axis=(0, 1))
axes[1].plot(spectral_bands, spec_train, "-o", color="#e67e22", label="Train (LV+PT)", ms=5)
axes[1].plot(spectral_bands, spec_test, "-o", color="#3498db", label="Test (EE)", ms=5)
axes[1].set_xlabel("Band"); axes[1].set_ylabel("Mean reflectance")
axes[1].set_title("Mean spectral profile: train vs test"); axes[1].legend()

plt.suptitle("Domain shift between LV+PT (train) and EE (test)", fontsize=13, y=1.02)
plt.tight_layout()
```

## Per-Country Data Availability

```python
fig, ax = plt.subplots(figsize=(10, 4))
for gi, g in enumerate(groups):
    mask = bench.groups == g
    avail = bench.s2_mask[mask].mean(axis=0)
    ax.plot(doy_mean, avail + gi * 0.02, "-", label=g, color=colors.get(g), linewidth=2)
ax.set_xlabel("Day of year"); ax.set_ylabel("Availability (offset)")
ax.set_title("Per-timestep data availability by country")
ax.legend()
plt.tight_layout()
```

## Cached Embeddings (if available)

```python
emb_path = REPO / "data" / "output" / "embeddings" / "eurocropsml" / "presto" / "n20000"
if emb_path.exists() and (emb_path / "clean.npy").exists():
    clean = np.load(emb_path / "clean.npy")
    print(f"Embedding shape: {clean.shape}")
    print(f"Statistics: mean={clean.mean():.4f}, std={clean.std():.4f}")

    from sklearn.decomposition import PCA
    pca = PCA(n_components=2, random_state=0)
    emb_2d = pca.fit_transform(clean[:5000])
    eg = bench.groups[:5000] if len(bench.groups) >= 5000 else bench.groups

    fig, ax = plt.subplots(figsize=(7, 5))
    for g in np.unique(eg):
        mask = eg == g
        ax.scatter(emb_2d[mask, 0], emb_2d[mask, 1],
                   c=colors.get(g, "#95a5a6"), s=5, alpha=0.5, label=g)
    ax.set_xlabel(f"PC1 ({pca.explained_variance_ratio_[0]:.1%})")
    ax.set_ylabel(f"PC2 ({pca.explained_variance_ratio_[1]:.1%})")
    ax.set_title("Presto embeddings (EuroCropsML, first 5k, PCA by country)")
    ax.legend(markerscale=3)
    plt.tight_layout()
else:
    print("No cached Presto embeddings found.")
```

## Summary

| Property | Value |
|----------|-------|
| Total parcels | 706,683 |
| Estonia | 175,906 (24.9%) |
| Latvia | 431,143 (61.0%) |
| Portugal | 99,634 (14.1%) |
| Time series length | 36–60 timesteps (variable) |
| Growing season | ~Feb–Nov 2021 |
| Spectral bands | 10 S2 bands + NDVI |
| Unique crop types | 93 HCAT codes |
| Split | LV+PT train → EE test |

**Key observations:**
- Massive scale: 706k parcels
- Significant domain shift: Southern Europe (PT) / Baltic (LV) → Northern Europe (EE)
- Estonia has fewer crop types and a shorter growing season
- ~10% of Estonian crop types unseen in training pool (unseen-class transfer)
- No S1 or climate data — S2-only
